```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('train.csv', sep = ',' )
test = pd.read_csv('test.csv', sep = ',')
```

```python
train.head()
```

```python
test.head()
```

```python
# On selectionne les variables qui me paraissent les plus importantes
# On nettoie ensuite 
train = train[['YearBuilt','TotalBsmtSF', '1stFlrSF','2ndFlrSF','OverallQual', 'BedroomAbvGr', 'KitchenAbvGr', 'TotRmsAbvGrd', 'GarageArea', 'PoolArea', 'MoSold', 'YrSold','SalePrice']]
```

```python
train_clean = train.dropna(axis=0, how='any')
```

```python
train_clean.head()
```

```python
Y_train = train_clean['SalePrice'].values
X_train = train_clean[['YearBuilt','TotalBsmtSF','1stFlrSF','2ndFlrSF','OverallQual','BedroomAbvGr','KitchenAbvGr','TotRmsAbvGrd','GarageArea','PoolArea','MoSold','YrSold']].as_matrix()
```

```python
test = test[['Id','YearBuilt','TotalBsmtSF', '1stFlrSF','2ndFlrSF','OverallQual', 'BedroomAbvGr', 'KitchenAbvGr', 'TotRmsAbvGrd', 'GarageArea', 'PoolArea', 'MoSold', 'YrSold']]
#X_testclean = test.dropna(axis=0, how='any')
X_testclean = test.fillna(0)
X_test = X_testclean.as_matrix()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.pairplot(train_clean)
plt.show()
```

```python
#repartition du prix des maisons
fig, ax = plt.subplots(figsize=(12,8))
sns.distplot(train_clean['SalePrice'], ax=ax)
plt.show()
```

```python
from sklearn import ensemble
from sklearn.utils import shuffle
from sklearn.metrics import mean_squared_error
# Test avec un Gradient Boosting regressor
clf = ensemble.GradientBoostingRegressor(n_estimators=150, max_depth=5)
clf.fit(X_train, Y_train)
```

```python
Y_predict = clf.predict(X_test[:,1:13])
```

```python
#Test avec un svm Kernel
from sklearn.svm import SVR
clf = SVR(kernel = 'rbf', gamma=0.1)
clf.fit(X_train, Y_train)
Y_predict = clf.predict(X_test[:,1:13])
```

```python
#Test avec un extra-trees
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.neighbors import KNeighborsRegressor
clf = ExtraTreesRegressor(n_estimators = 240, bootstrap=True, oob_score=True)
clf.fit(X_train, Y_train)
Y_predict = clf.predict(X_test[:,1:13])
```

```python
# gridsearch pour dÃ©terminer les meilleurs paramÃªtres
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
ExtraT = ExtraTreesRegressor()
print(cross_val_score(ExtraT, X_train, Y_train, cv=10))
```

```python
parameters = {'n_estimators':[30,50,100,150,200,240,280,300,400],'bootstrap':(True, True), 'oob_score':(True, False)}
clf = GridSearchCV(ExtraT, parameters, cv=10)
clf.fit(X_train, Y_train)
clf.best_params_
```

```python
#Test rigolo avec un KNN
clf = KNeighborsRegressor()
neigh = KNeighborsRegressor(n_neighbors=10)
neigh.fit(X_train, Y_train)
Y_predict = neigh.predict(X_test[:,1:13])
```

```python
#Regression Lasso
from sklearn import linear_model
clf = linear_model.Lasso(alpha=0.1)
clf.fit(X_train,Y_train)
Y_predict = clf.predict(X_test[:,1:13])
```

```python
# Test avec un Random Forest
from sklearn.ensemble import RandomForestRegressor
regr = RandomForestRegressor(n_estimators=240, max_depth=10, random_state=0, bootstrap=True, oob_score=True)
regr.fit(X_train, Y_train)
Y_predict = regr.predict(X_test[:,1:13])
```

```python
#import au format CSV
import csv
Y_predict = pd.DataFrame(Y_predict)
Y_predict.columns = ['SalePrice']
df1 = Y_predict
df2 = X_testclean[['Id']]
Prix_Maisons = pd.concat([df2, df1], axis = 1)
Prix_Maisons.to_csv('PrixMaisons.csv', index=False, sep=',')
```

```python
#On rajoute quelques variables : LotArea, OverallCond
train = train[['YearBuilt','LotArea','OverallCond','TotalBsmtSF', '1stFlrSF','2ndFlrSF','OverallQual', 'BedroomAbvGr', 'KitchenAbvGr', 'TotRmsAbvGrd', 'GarageArea', 'PoolArea', 'MoSold', 'YrSold','SalePrice']]
```

```python
train_clean = train.dropna(axis=0, how='any')
```

```python
Y_train = train_clean['SalePrice'].values
X_train = train_clean[['YearBuilt','LotArea','OverallCond','TotalBsmtSF','1stFlrSF','2ndFlrSF','OverallQual','BedroomAbvGr','KitchenAbvGr','TotRmsAbvGrd','GarageArea','PoolArea','MoSold','YrSold']].as_matrix()
test = test[['Id','YearBuilt','LotArea','OverallCond','TotalBsmtSF', '1stFlrSF','2ndFlrSF','OverallQual', 'BedroomAbvGr', 'KitchenAbvGr', 'TotRmsAbvGrd', 'GarageArea', 'PoolArea', 'MoSold', 'YrSold']]

X_testclean = test.fillna(0)
X_test = X_testclean.as_matrix()
```

```python
# Test avec un Random Forest
from sklearn.ensemble import RandomForestRegressor
regr = RandomForestRegressor(n_estimators=240, max_depth=10, random_state=0, bootstrap=True, oob_score=True)
regr.fit(X_train, Y_train)
Y_predict = regr.predict(X_test[:,1:15])
```

```python
#Test avec un ExtraTrees
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
ExtraT = ExtraTreesRegressor()
print(cross_val_score(ExtraT, X_train, Y_train, cv=10))
```

```python
clf = ExtraTreesRegressor(n_estimators = 240, bootstrap=True, oob_score=True)
clf.fit(X_train, Y_train)
Y_predict = clf.predict(X_test[:,1:15])
```

```python
#PolynomialFeatures + ExtraTrees
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(4)
poly4 = poly.fit_transform(X_train)
```

```python
#poly.get_feature_names()
from sklearn.ensemble import ExtraTreesRegressor
clf = ExtraTreesRegressor(n_estimators = 240, bootstrap=True, oob_score=True)
clf.fit(poly4, Y_train)
Y_predict = clf.predict(poly.fit_transform(X_test[:,1:15]))
```

```python
#Test avec un Random Forest
from sklearn.ensemble import RandomForestRegressor
regr = RandomForestRegressor(n_estimators=240, max_depth=10, random_state=0, bootstrap=True, oob_score=True)
regr.fit(poly4, Y_train)
Y_predict = regr.predict(poly.fit_transform(X_test[:,1:15]))
```

```python
import csv
Y_predict = pd.DataFrame(Y_predict)
Y_predict.columns = ['SalePrice']
df1 = Y_predict
df2 = X_testclean[['Id']]
Prix_Maisons = pd.concat([df2, df1], axis = 1)
Prix_Maisons.to_csv('PrixMaisons.csv', index=False, sep=',')
```
